```python
import pandas as pd
df=pd.read_csv('spam_ham_dataset.csv')
df.head()
```

```python
input=df.drop(['Unnamed: 0'],axis='columns')
input
```

```python
input.groupby('label').describe()
input = input.drop('label_num',axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
input['spam']=LabelEncoder().fit_transform(input['label'])
input
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(input.text,input.spam,test_size=0.1)
len(x_test)
len(x_train)
```

**Using CountVectorizer for naming the words in x_test**

```python
from sklearn.feature_extraction.text import CountVectorizer

# Fit on training data
v = CountVectorizer()
x_train_new = v.fit_transform(x_train.values)
print(x_train_new.toarray()[:3])  # View the first 3 samples in array format

# Transform test data using the same vocabulary
x_test_new = v.transform(x_test.values)
print(x_test_new.toarray()[:3])  # View the first 3 samples in array format
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
model.fit(x_train_new,y_train)
```

```python
model.score(x_test_new,y_test)
```
